```python
#implementing a low pass filter 1)averaging method
import numpy as np
from matplotlib import pyplot as mplt
img=cv.imread('cat.png')

blur=cv.blur(img,(5,5))

mplt.subplot(121)
mplt.imshow(img)
mplt.title('original')
mplt.xticks([]),mplt.yticks([])

mplt.subplot(122)
mplt.imshow(blur)
mplt.title('blurred_with_avg')
mplt.xticks([]),mplt.yticks([])

mplt.show()
```

```python

blur=cv.GaussianBlur(img,(5,5),0)

mplt.subplot(121)
mplt.imshow(img)
mplt.title('original')
mplt.xticks([]),mplt.yticks([])

mplt.subplot(122)
mplt.imshow(blur)
mplt.title('blurred_with_avg')
mplt.xticks([]),mplt.yticks([])

mplt.show()
```

```python

blur=cv.medianBlur(img,11)

mplt.subplot(121)
mplt.imshow(img)
mplt.title('original')
mplt.xticks([]),mplt.yticks([])

mplt.subplot(122)
mplt.imshow(blur)
mplt.title('blurred_with_avg')
mplt.xticks([]),mplt.yticks([])

mplt.show()
```

```python
pimple=cv.imread('pimple.png')
blur=cv.bilateralFilter(pimple,21,75,75)

mplt.subplot(121)
mplt.imshow(pimple)
mplt.title('original')
mplt.xticks([]),mplt.yticks([])

mplt.subplot(122)
mplt.imshow(blur)
mplt.title('blurred_with_bilateral')
mplt.xticks([]),mplt.yticks([])

mplt.show()
```

```python
# filter to implement image sharpening
kernel=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])

blur=cv.filter2D(img,-1,kernel)

mplt.subplot(121)
mplt.imshow(img)
mplt.title('original')
mplt.xticks([]),mplt.yticks([])

mplt.subplot(122)
mplt.imshow(blur)
mplt.title('sharpen')
mplt.xticks([]),mplt.yticks([])

mplt.show()
```

```python
# filter to implement edge detection Type 1 kernel
kernel=np.array([[1,0,-1],[0,0,0],[-1,0,1]])

blur=cv.filter2D(img,-1,kernel)

mplt.subplot(121)
mplt.imshow(img)
mplt.title('original')
mplt.xticks([]),mplt.yticks([])

mplt.subplot(122)
mplt.imshow(blur)
mplt.title('edgeDection')
mplt.xticks([]),mplt.yticks([])

mplt.show()
```

```python
# filter to implement edge detection with Type 2 kernel
kernel=np.array([[0,1,0],[1,-4,1],[0,1,0]])

blur=cv.filter2D(img,-1,kernel)

mplt.subplot(121)
mplt.imshow(img)
mplt.title('original')
mplt.xticks([]),mplt.yticks([])

mplt.subplot(122)
mplt.imshow(blur)
mplt.title('edgeDection')
mplt.xticks([]),mplt.yticks([])

mplt.show()
```

```python
# filter to implement edge detection with Type 3 kernel
kernel=np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])

blur=cv.filter2D(img,-1,kernel)

mplt.subplot(121)
mplt.imshow(img)
mplt.title('original')
mplt.xticks([]),mplt.yticks([])

mplt.subplot(122)
mplt.imshow(blur)
mplt.title('edgeDection')
mplt.xticks([]),mplt.yticks([])

mplt.show()
```

```python
# video capturing in opencv
import numpy as np
import cv2 
cap = cv2.VideoCapture(0)
while(True):
# Capture frame-by-frame
  ret, frame = cap.read()
# Our operations on the frame come here
  gray = cv2.cvtColor(frame,1)
# Display the resulting frame
  cv2.imshow('frame',gray)
  if cv2.waitKey(1) & 0xFF == ord('q'):
      break
# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```
